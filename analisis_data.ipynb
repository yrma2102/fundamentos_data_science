#Ejercicio 1 Limpieza de datos con Python:

```python
#importando lo necesario
import pandas as pd
```

```python
#importando archivo
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
#mosrtrar los primeros registros
df.head()
```

```python
#Identificar y solucionar cualquier valor duplicado.
df.info()
#segun lo revisado la cantidad de registros es de 7043 para todas las columnas no hay datos nulos
```

```python
cantidad_duplicados = df.duplicated().sum()
print(cantidad_duplicados)
```

```python
# segun analisis no hay datos duplicados
df = df.drop_duplicates()
df.info()
#siguen existiendo las 7043 filas
```

```python
#Asegurarse de que todas las columnas coincidan con los tipos de datos indicados en el diccionario de datos.
df.describe()
```

```python
#segun lo que se puede observar al aplicar un df.info() y df.describe() faltaria covertir la columna Total Charges a float

#limpiamos y verificamos que existan vacios
df['TotalCharges'] = df['TotalCharges'].str.strip()
df['TotalCharges'].replace('', pd.NA, inplace=True) #convierte cadenas vacÃ­as en NA
df.info()
```

```python
#despues de  la limpieza se visualiza que faltan datos en TotalCharges
#vemos los meses que tiene el usuario con el cargo mensual para saber si se puede agregar el producto de esto
df[['tenure', 'MonthlyCharges', 'TotalCharges']]
```

```python
# se puede ver que el cargoTotal es el producto de los meses que tiene el usuario y el cargo mensual
#con esa informacion podemos completar la columna TotalCharges
#crear una columna de TotalChargeNumeric para aplicar el calculo a los valores vacios

df['TotalCharges_numeric'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.info()
```

```python
#aplicamos el calculo de Ternure * MontlyCharges a aquellos valores NAN
#usamos loc para aplicar el calculo solo a las filas con NAN
df.loc[df['TotalCharges_numeric'].isna(), 'TotalCharges_numeric'] = (
    df['tenure'] * df['MonthlyCharges']
)
df.info()
```

```python
#Asegurarse de que todas las columnas coincidan con los tipos de datos indicados en el diccionario de datos.

df.describe()
#aplicando la limpieza de  la columna TotalCharges y creando nueva columna de tipo Float TotalCharges_numeric agregando los datos faltantes
#se aplica el df.desribe para tener una estadistica de los columnas numericas.
```

```python
# se agrega una columna para categorizar al clientes de acuerdo al cargo mensual que pagan
def categorizar_pago(pago):
    #18-118
    if pago < 30:
        return 'Bajo'
    elif 25 <= pago < 90:
        return 'Medio'
    else:
        return 'Alto'

df['Categoria_Pago'] = df['MonthlyCharges'].apply(categorizar_pago)
# se usa apply para llamar una funcion que clasifica el rango de pago
df.groupby('Categoria_Pago').size()
```

#Ejercicio 2 ExploraciÃ³n de datos con Python:

```python
#Crear dos tipos diferentes de visualizaciones exploratorias univariadas.
#  Cada visualizaciÃ³n debe incluir una breve interpretaciÃ³n dentro del archivo de cÃ³digo.
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df['MonthlyCharges'], bins=8, color='blue', edgecolor='black')
plt.title('DistribuciÃ³n de MonthlyCharges')
plt.xlabel('MonthlyCharges')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

#InterpretaciÃ³n
#El cargo mensual que pagan los clientes minimo es 18 y maximo 118
#Del grafico de histograma se puede apreciar que hay muchos clientes que pagan entre 20 y 30 $
#y hay menos clientes que pagan entre 30 y 50 $
```

```python


# Creamos el boxplot.
plt.boxplot(df['tenure'])
plt.title('Diagrama de permanencia en meses')
plt.ylabel('Meses')
plt.grid(True)
plt.show()

#InterpretaciÃ³n
#La mayorÃ­a de los clientes permanecen  entre 10 y 55 meses.
# La mediana esta alrededor de 29 meses
# No e observan outliers hacia el extremo alto ni bajo
```

```python
#Crear dos tipos diferentes de visualizaciones exploratorias multivariadas.
#  Cada visualizaciÃ³n debe incluir una breve interpretaciÃ³n dentro del archivo de cÃ³digo.
plt.figure(figsize=(8, 6))
plt.scatter(df['tenure'], df['TotalCharges_numeric'], alpha=0.5, color='yellow')
plt.title('RelaciÃ³n entre Meses y Gasto Total')
plt.xlabel('Meses')
plt.ylabel('Gasto Total')
plt.grid(True)
plt.show()

# InterpretaciÃ³n:
# el grafico muestra que a mayor cantidad de meses que tiene el cliente en la compaÃ±ia mas alto es su gatos total.
```

```python
#crea una Serie donde el indice es el tipo de contrato con su respectiva lista de valores de cargos del mes
grupos = df.groupby('Contract')['MonthlyCharges'].apply(list)

# Etiquetas con los tipos de contrato
labels = df['Contract'].unique()

# Crear el boxplot
plt.boxplot(grupos, labels=labels)
plt.title('ComparaciÃ³n entre los gastos mensuales por tipo de contrato')
plt.ylabel('Monthly Charges')

plt.show()

#InterpretaciÃ³n
#segun el grafico se observa si el contrato es a dos aÃ±os tiene una caja mas baja que indica que pagan cuotas mas bajas
#Tambien se observa que en los contrato de mes a mes la caja es mas alta lo que indica que pagan cuotas mas altas
#En los contractos de one aÃ±o la caja es bastante similar al contrato de dos aÃ±os
```
